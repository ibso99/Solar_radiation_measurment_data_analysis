```python
print("Lets code together")
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
# 1. Benin Solar Radiation Measurement Exploratory Data analysis  
# starting from fetching the data from local repo!
file_path = "C:\\Users\\ibsan\\Desktop\\week_0\\Data_set\\data\\data\\benin-malanville.csv"
benin_dataset = pd.read_csv(file_path, sep=',')
benin_dataset.head()
```

```python
print(benin_dataset.info())
print(benin_dataset.describe()) 
```

```python
###############################
# checking the overall data infformation including data types....
print(benin_dataset.info())
###########################
#  checking the shape of the datasets i.e the number of rows and columns 
print(benin_dataset.shape)
```

```python
# seeing if there is any null values in the entire dataset
print(benin_dataset.isnull().sum())
```

```python
# visualizing missing data

sns.heatmap(benin_dataset.isnull(), cbar=False, cmap="viridis")
plt.title("Missing Values Heatmap")
plt.show()
```

```python
# Understanding the distribution of the variable GHI and GNI from the datasets

# histogram and density plot
for col in ["GHI","DNI"]:
    sns.histplot(benin_dataset[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Identifying outliers using Box-plot

sns.boxplot(x=benin_dataset['GHI'])
plt.title('Boxplot of GHI')
plt.show() 


# Detecting outliers using Inter quirtile range (IQR)

Q1 = benin_dataset['GHI'].quantile(0.25)
Q2 = benin_dataset['GHI'].quantile(0.75)
print(f'IQR is {Q1-Q2}')
```

```python
# Checking skewness of the two variables 
from scipy.stats import skew 
print(f"Skewness of the GHI: {skew(benin_dataset['GHI'])}")
print(f"Skewness of the DNI: {skew(benin_dataset['DNI'])}")
```

```python
print(benin_dataset.columns)
```

```python
# Exploring the relationship between the GHI variable and DNI variable 
# employing scatter-plot, correlation matrix and pair plot 

######################################################################
# 1. Correlation matrix 
corr = benin_dataset.iloc[:,1:].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
# Exploring the relationship between the GHI variable and DNI variable 

######################################################################

# 2. Scatter-plot 
sns.scatterplot(x='GHI', y='DNI', data=benin_dataset)
plt.title("Relationship between GHI and GNI")
plt.xlabel('GHI')
plt.ylabel('GNI')
plt.show()

```

```python
######################################################################
# 3. pairplot for multiple variables i.e matrix of scatter-plot

sns.pairplot(benin_dataset[['GHI', 'DNI', 'DHI']]) 
plt.show()
```
